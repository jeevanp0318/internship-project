```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model  import LinearRegression, LassoCV, Lasso, MultiTaskLassoCV
from sklearn.multioutput import MultiOutputRegressor

pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("Preprocessed_Bondora.csv")
```

```python
data
```

```python
#data.drop(['Target',],axis = 1, inplace = True)
```

```python
X = data.drop(['Amount','Interest','LoanDuration'] ,axis=1)
y = data[['Amount','Interest','LoanDuration']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
print("Training Data Shape =", X_train.shape)
print("Testing Data Shape =", X_test.shape)
```

```python
X_train
```

```python
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

```python
X_train
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = X_train.values
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables, i) for i in range(variables.shape[1])]
vif["Features"] = X_train.columns
```

```python
vif
```

```python
from sklearn import metrics
def get_train_test_rmse(model):
    y_train_pred = model.predict(X_train)
    rmse_train = np.sqrt(metrics.mean_squared_error(y_train, y_train_pred))
    
    y_test_pred = model.predict(X_test)
    rmse_test = np.sqrt(metrics.mean_squared_error(y_test,y_test_pred))
    
    print("train: ", rmse_train, "test: ",rmse_test)
```

```python
lassocv= MultiTaskLassoCV(alphas=[0.025,0,0.050,0.080, 0.004,0.50,0.25,1,10], cv=10, max_iter=1000, random_state=24)
lassocv.fit(X_train,y_train)
```

```python
alpha = lassocv.alpha_
alpha
```

```python
lasso_reg = Lasso(alpha=alpha)
model = MultiOutputRegressor(lasso_reg)
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train), model.score(X_test,y_test)
```

```python
get_train_test_rmse(model)
```

```python
linear = LinearRegression()
multi = MultiOutputRegressor(linear)
multi.fit(X_train,y_train)
```

```python
multi.score(X_train,y_train)
```

```python
multi.score(X_test,y_test)
```
